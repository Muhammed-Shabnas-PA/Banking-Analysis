### Risk category by income band

```python
# Convert 'Estimated Income' to numeric (handle non-numeric gracefully)
df['Estimated Income'] = pd.to_numeric(df['Estimated Income'], errors='coerce')

# Define income bins and labels
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Mid', 'High']

# Create income band column
df['Income_Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)

# Count the number of customers in each income band
band_counts = df['Income_Band'].value_counts().sort_index()

# Plot the income band distribution
plt.figure(figsize=(8, 6))
band_counts.plot(kind='bar', color=['skyblue', 'orange', 'green'])
plt.title('Income Band Distribution')
plt.xlabel('Income Band')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

**Observations:**
- The bar plot displays the number of customers in each income band: Low, Mid, and High.
- The "High" income band has the largest number of customers, followed by "Mid" and then "Low".
- There is a noticeable skew toward the "High" income band, indicating a concentration of customers with higher estimated incomes.

**Overall Insights:**
- The bank's customer base is predominantly composed of individuals in the "High" income band.
- This distribution suggests that the bank may attract or retain more high-income clients, which could influence its product offerings and marketing strategies.
- Understanding the income band distribution is important for risk assessment, as higher-income customers may have different financial needs and risk profiles compared to lower-income segments.

```python
# Plot histogram with KDE for Estimated Income
plt.figure(figsize=(8, 5))
sns.histplot(df['Estimated Income'], bins=30, kde=True, color='orange')
plt.title('Estimated Income Distribution')
plt.xlabel('Estimated Income')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Calculate and print skewness
income_skewness = df['Estimated Income'].skew()
print(f"Skewness of Estimated Income: {income_skewness:.2f}")
```

- The distribution has a long right tail, indicating the presence of high-income outliers.
- The majority of customers fall into the "Mid" and "High" income bands, as seen in the earlier band_counts summary.

**Overall Insights:**
- The right-skewed nature of the income distribution is typical in financial datasets, where a small proportion of customers have significantly higher incomes.
- Identifying high-income outliers is important for targeted marketing, premium product offerings, and risk segmentation.
- The skewness value (~0.88) confirms the presence of more lower-income customers and a few high-income individuals.
- Understanding this distribution helps the bank tailor its services and risk management strategies to different customer segments, ensuring both inclusivity and profitability.

### GenderId Count Plot

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='GenderId', data=df, palette='pastel')
plt.title('Customer Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')

# Set custom x-tick labels: 1 = Male, 2 = Female
ax.set_xticklabels(['Male', 'Female'])

# Annotate each bar with the count value
for p in ax.patches:
    count = int(p.get_height())
    ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2, count), 
                ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()
```

**Observations:**
- The count plot displays the number of customers by gender, with separate bars for Male and Female.
- The heights of the bars indicate the total number of male and female customers in the dataset.
- Each bar is annotated with the exact count value for clarity.

**Overall Insights:**
- The plot allows for a quick comparison of the gender distribution within the customer base.
- If one gender significantly outnumbers the other, it may suggest a demographic skew that could impact marketing, product development, or risk assessment strategies.
- A balanced distribution indicates that the bank appeals equally to both genders, while an imbalance may highlight opportunities for targeted outreach or further investigation into underlying causes.
